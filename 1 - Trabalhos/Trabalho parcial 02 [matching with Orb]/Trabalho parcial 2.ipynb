```
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt

# Carregar imagens
spider = cv.imread('spiderperto.jpg',cv.IMREAD_GRAYSCALE)  
spidercompleto = cv.imread('spidercompleto.jpg',cv.IMREAD_GRAYSCALE)

# Inicializa o ORB
orb = cv.ORB_create()

# Encontra os pontos chaves e suas descrições
chave1, descricao1 = orb.detectAndCompute(spider,None)
chave2, descricao2 = orb.detectAndCompute(spidercompleto,None)

# Cria o objeto BFMatcher
bf = cv.BFMatcher(cv.NORM_HAMMING, crossCheck=True)

# Combina descrições
matches = bf.match(descricao1,descricao2)

# Organiza as combinações em ordem de distância
matches = sorted(matches, key = lambda x:x.distance)

# Desenha as primeiras combinações, definidadas entre chaves
final = cv.drawMatches(spider,chave1,spidercompleto,chave2,matches[:50],None,flags=cv.DrawMatchesFlags_NOT_DRAW_SINGLE_POINTS)

# printa imagem final
plt.imshow(final),plt.show()
```
